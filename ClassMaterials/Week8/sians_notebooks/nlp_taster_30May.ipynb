## get chosen text

```python
mytext=docx2txt.process('Horseracing.docx')
```

```python
mytext
```

## nltk tools to process the text 

```python
from nltk.tokenize import word_tokenize
```

```python
mytext=word_tokenize(mytext)
```

```python
len(mytext)
```

```python
# lets put everything in lower case 
lower_text=[word.lower() for word in mytext]
```

```python
# lets count the frequency of words - first version 
fdist_v1=nltk.FreqDist(lower_text)
```

```python
word_freq_v1=dict((word,freq) for word, freq in fdist_v1.items())
```

```python
# sort it by the value freq 
sorted(word_freq_v1.items(), key=lambda x:x[1], reverse=True)
```

## text cleaning - depends what you are trying to achieve

- remove stop words
- remove punctuation 
- lemmatising / stemming 
- pronouns, nouns and entities/ abbreviations 
- prepositions 

```python
# punctuation spaces numbers - isalpha() - apply to lower_text 
clean_text=[word for word in lower_text if word.isalpha()]
```

```python
len(clean_text)
```

```python
# get stop words - appropriate for text 
from nltk.corpus import stopwords
```

```python
my_stopwords=stopwords.words('english')
```

```python
my_stopwords
```

```python
print(stopwords.fileids())
```

```python
clean_text=[word for word in clean_text if word not in my_stopwords]
```

```python
len(clean_text)
```

```python
# lets count the frequency of words - second version 
fdist_v2=nltk.FreqDist(clean_text)
word_freq_v2=dict((word,freq) for word, freq in fdist_v2.items())
sorted(word_freq_v2.items(), key=lambda x:x[1], reverse=True)
```

```python
porter = nltk.PorterStemmer()
```

```python
stem_words=[porter.stem(word) for word in clean_text]
```

```python
#third version of word frequency 
fdist_v3=nltk.FreqDist(stem_words)
word_freq_v3=dict((word,freq) for word, freq in fdist_v3.items())
sorted(word_freq_v3.items(), key=lambda x:x[1], reverse=True)
```

## get sentiment scores 

```python
# textblob - cant handle nulls 
clothingdf=clothingdf.dropna(subset=['Review Text'], how='any')
```

```python
clothingdf.info()
```

```python
clothingdf.iloc[34]['Review Text']
```

```python
clothingdf.iloc[34]['Rating']
```

```python
example=TextBlob(clothingdf.iloc[34]['Review Text'])
example.sentiment
```

```python
def get_sentiment_p(text):
    return TextBlob(text).sentiment.polarity
```

```python
clothingdf['sentiment_p']=clothingdf['Review Text'].apply(get_sentiment_p)
```

```python
def get_sentiment_s(text):
    return TextBlob(text).sentiment.subjectivity
```

```python
clothingdf['subjectivity']=clothingdf['Review Text'].apply(get_sentiment_s)
```

```python
clothingdf.head()
```

## analysis  - avg rating, sentiment, subjectivity  per dept div 

```python
sent_df=clothingdf.groupby(['Division Name','Department Name']).agg({'Rating':'mean','sentiment_p':'mean', 'subjectivity':'mean'})
```

```python
sent_df.columns
```

```python
clothingdf['dep_div']=clothingdf['Division Name']+' - '+clothingdf['Department Name']
```

```python
sent_df=clothingdf.groupby(['dep_div']).agg({'Rating':'mean','sentiment_p':['mean','min','max'], 'subjectivity':'mean'})
```

```python
sent_df
```

```python
tokens=nlp(''.join(str(tweets.text.tolist())))
```

```python
len(tokens)
```

```python
tokens[0:6]
```

```python
from collections import Counter
```

```python
items=[x.text for x in tokens.ents] # finds all entities , ignores stopwords and punctuation
Counter(items).most_common(20)
```

```python
person_list=[]
for ent in tokens.ents:
    if ent.label_=='PERSON':
        person_list.append(ent.text)
```

```python
person_list
```

```python
person_counts=Counter(person_list).most_common(20)
df_person=pd.DataFrame(person_counts,columns=['text','count'])
df_person.plot.barh(x='text',y='count');
```
